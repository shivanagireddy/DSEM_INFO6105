
##Abstract:

This report presents an analysis of a dataset containing information about various cars, including attributes like car name, year of manufacture, selling price, present price, kilometers driven, fuel type, seller type, transmission, and owner count. The aim of this analysis is to develop a predictive model to estimate the selling price of cars based on their features.

The report encompasses four main stages: data cleaning, feature selection, modeling, and interpretability. In the data cleaning phase, we address missing values, inconsistencies, and outliers to ensure data quality. Feature selection involves identifying the most relevant features that contribute to the prediction of selling price.

For modeling, we employ machine learning algorithms to build a predictive model, evaluating and selecting the most suitable one based on performance metrics. Finally, interpretability techniques are applied to explain the model's predictions and understand the factors influencing the selling price of cars.

Through this comprehensive analysis, we aim to provide valuable insights into the factors affecting car prices and develop a reliable model for predicting selling prices, aiding both sellers and buyers in making informed decisions in the automotive market.

```python
# Encoding categorical variables

import pandas as pd

# Sample data
data = {
    "Seller_Type": ["Dealer", "Individual"],
    "Transmission": ["Manual", "Automatic"],
    "Fuel_Type": ["Petrol", "Diesel"]
}

# Create a DataFrame from the data
car_df = clean_df.copy()

# Define encoding functions (same as before)
def binary_encode(category):
  if category == "Dealer" or category == "Manual":
    return 0
  else:
    return 1

fuel_type_mapping = {"Petrol": 1, "Diesel": 2, "CNG": 3}

# Encode Seller_Type and Transmission (using apply)
car_df["Seller_Type_Encoded"] = car_df["Seller_Type"].apply(binary_encode)
car_df["Transmission_Encoded"] = car_df["Transmission"].apply(binary_encode)

# Encode Fuel_Type (using lookup)
car_df["Fuel_Type_Encoded"] = car_df["Fuel_Type"].replace(fuel_type_mapping)

# Remove original categorical columns
car_df.drop(columns=["Seller_Type", "Transmission", "Fuel_Type"], inplace=True)

# Print results (shows the encoded DataFrame)
print(car_df)
```

* The DataFrame 'clean_df' is copied into another DataFrame named 'car_df'. This is done to preserve the original DataFrame while performing encoding.

* Encoding is applied to the 'Seller_Type' and 'Transmission' columns using the apply method:
* The binary_encode function is applied to each value in the 'Seller_Type' and 'Transmission' columns, and the results are stored in new columns named 'Seller_Type_Encoded' and 'Transmission_Encoded', respectively.
* Encoding of 'Fuel_Type' is applied using the replace method:
The dictionary fuel_type_mapping is used to replace fuel types with their corresponding numerical values, and the result is stored in a new column named 'Fuel_Type_Encoded'.

```python
# Defining the features

features = ['Present_Price' , 'Fuel_Type_Encoded', 'Transmission_Encoded']

target = ['Selling_Price']

X = car_df[features]

y = car_df[target]

# Split the data into a training set and a testing set
# test_size specifies the proportion of the data to be included in the test split
# random_state ensures reproducibility of your split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1111)


# Check the size of the splits
print(f'Training set size: {X_train.shape[0]} samples')
print(f'Test set size: {X_test.shape[0]} samples')
```

* The test_size parameter specifies the proportion of the data to include in the testing set. Here, it's set to 0.2, which means 20% of the data will be used for testing.

* The random_state parameter ensures reproducibility of the split. Setting it to a specific value (1111 in this case) ensures that the same split is obtained every time the code is run.

* There are 239 samples in the training set and 60 samples in the test set.

##Training the model on linear model

```python
# Adding a constant to the predictors because statsmodels' OLS doesn't include it by default
X_train_const = sm.add_constant(X_train)
X_train_const
```

```python
# Fit the OLS model
model_fitted = sm.OLS(y_train, X_train_const).fit()

# Printing Summary
print(model_fitted.summary())
```

* the output suggests that the OLS model is a statistically significant fit for the data. The Present_Price and Fuel_Type_Encoded features are likely important predictors of Selling Price. However, it's important to note that the constant term is negative and doesn't make sense in the context of car prices. This might indicate that there are other important factors missing from the model that are affecting Selling Price. You can also explore other metrics like R-squared to see how well the model generalizes to unseen data.

##Prediction and Testing

```python
# Adding a constant to the test predictors
X_test_const = sm.add_constant(X_test)

# Making predictions on the test set
test_predictions = model_fitted.predict(X_test_const)
test_predictions
```

* The output is a pandas Series containing the predicted selling prices for each observation in the test set.
* Each row corresponds to a specific observation in the test set.
* The predicted selling prices are represented as floating-point numbers.

* the output represents the model's predictions for selling prices based on the features provided in the test set, which can be used for evaluation and decision-making purposes in various applications.

```python
# Scatter plot for observed vs predicted values on test data
plt.scatter(y_test, test_predictions, color = "forestgreen")
plt.xlabel('Observed Values')
plt.ylabel('Predicted Values')
plt.title('Observed vs Predicted Values on Test Data')
plt.plot(y_test, y_test, color='darkred')  # line for perfect prediction (true values)
plt.show()
```

* The data points are colored forest green. The dark red line represents a perfect prediction, where the predicted values would exactly match the observed values. Ideally, most of the data points should fall near this line.

* the scatter plot suggests that the linear regression model is able to capture some of the relationship between the features (Present_Price, Fuel_Type_Encoded) and Selling Price, but there is also room for improvement. You might want to explore other features or modeling techniques to see if you can improve the accuracy of the model.

```python
# Calculate the mean of the residuals
mean_residuals = np.mean(model_fitted.resid)

print(f"The mean of the residuals is {np.round(mean_residuals,2)}")
```

```python
# Plotting the residuals
plt.scatter(model_fitted.fittedvalues, model_fitted.resid, color = "forestgreen")
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()
```

* Ideally, the residuals would be randomly scattered around a horizontal line at zero. This would indicate that the model's errors (residuals) are independent of the fitted values, and that the variance of the errors is constant across the range of predictions.

* the residual plot suggests that the linear regression model meets some of the assumptions for linear regression, which is a good thing. However, it's always a good practice to consider other diagnostics and performance metrics to assess the model's quality.

```python
# Calculate the residuals
residuals = model_fitted.resid

# Check for correlation between residuals and each predictor
for column in X_train.columns:
    corr_coefficient = np.corrcoef(X_train[column], residuals)[0, 1]
    print(f'Correlation between residuals and {column}: {np.round(corr_coefficient,2)}')
```

```python
# Plotting the residuals
plt.scatter(model_fitted.fittedvalues, model_fitted.resid, color = "forestgreen")
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

* For each value of max_leaf_nodes (5, 50, 500, 5000), the code prints the corresponding MAE value.
* The MAE values indicate the average absolute difference between the actual and predicted target values on the validation set for different configurations of the decision tree model.
* In this case, all the MAE values are 1, regardless of the value of max_leaf_nodes.
* This suggests that increasing the max_leaf_nodes parameter beyond a certain point does not significantly improve model performance, at least on this validation set.

##Data Preprocessing

```python
data['Selling_Price'] = data['Selling_Price'].map(lambda x: re.sub(r"[^0-9.-]", "", str(x)))
data['Kms_Driven'] = data['Kms_Driven'].map(lambda x: re.sub(r"[^0-9.-]", "", str(x)))
data['Present_Price'] = data['Present_Price'].map(lambda x: re.sub(r"[^0-9.-]", "", str(x)))

data['Selling_Price'] = pd.to_numeric(data['Selling_Price'])
data['Kms_Driven'] = pd.to_numeric(data['Kms_Driven'])
data['Present_Price'] = pd.to_numeric(data['Present_Price'])
data=data.dropna()

data_pd=data.copy()

data = h2o.H2OFrame(data)
```

this code snippet performs data cleaning by removing non-numeric characters from specific columns using regular expressions, converts data types to numeric, drops rows with missing values, creates a copy of the processed data in Pandas DataFrame format, and then converts the data into an H2OFrame for further analysis using H2O's machine learning algorithms.

```python
import seaborn as sns
f,ax = plt.subplots(figsize=(5, 5))
mask = np.zeros_like(data_pd.corr().corr())
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data_pd.corr(), annot=True, linewidths=0.4,linecolor="white", fmt= '.1f',ax=ax,cmap="Blues",mask=mask)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
df_vif = data_pd.drop(columns={"Car_Name","Transmission","Seller_Type"})
# df_vif.drop(columns={'song_popularity','song_name'}, inplace = True)

X = df_vif.loc[:,df_vif.columns != 'Fuel_Type']

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] =X.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

* The output displays the feature names along with their corresponding VIF values.
* For example, the feature "Year" has a VIF of approximately 2.78, indicating a low level of multicollinearity.
* Conversely, both "Selling_Price" and "Present_Price" have VIF values around 9.35, suggesting a higher degree of multicollinearity between these features.
* "Kms_Driven" and "Owner" have relatively low VIF values, indicating less multicollinearity compared to other features.
